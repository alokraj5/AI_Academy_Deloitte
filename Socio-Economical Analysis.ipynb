Cylical employment variation is essentially the variation of how many people are being hired or fired due to the shifts in the conditions of the economy.

```python
emp_var_status=pd.pivot_table(data=cust_sedata,index=['emp.var.rate'],columns=['y'],values=['Customer_id'],aggfunc='count',margins=True, margins_name='Total')
emp_var_status['Success Rate']=(emp_var_status['Customer_id']['yes']/emp_var_status['Customer_id']['Total'])*100
emp_var_status.sort_values(by='Success Rate',ascending=False,inplace=True)
emp_var_status
```
